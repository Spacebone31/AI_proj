```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from keras.models import Sequential, Model
from keras.layers import Input, Add, Conv1D, MaxPooling1D, Flatten, LSTM, GRU, Dense, SimpleRNN,ReLU,DepthwiseConv1D,GlobalAveragePooling1D,BatchNormalization,Layer
from statsmodels.tsa.statespace.sarimax import SARIMAX
import keras.backend as K
import tensorflow as tf
from scipy.optimize import minimize

# Load the datasets
data1 = pd.read_csv('jogja2023.csv', delimiter=';')
data2 = pd.read_csv('jogja2022.csv', delimiter=';')
data3 = pd.read_csv('jogja2021.csv', delimiter=';')
data4 = pd.read_csv('jogja2020.csv')

# Preprocess the data (assuming both datasets have the same structure)
data1['Date'] = pd.to_datetime(data1['Waktu'], format='%d/%m/%Y', errors='coerce')
data2['Date'] = pd.to_datetime(data2['Waktu'], format='%d/%m/%Y', errors='coerce')
data3['Date'] = pd.to_datetime(data3['Waktu'], format='%d/%m/%Y', errors='coerce')
data4['Date'] = pd.to_datetime(data4['Date'], format='%m/%d/%Y', errors='coerce')

# Drop rows with invalid dates
data1.dropna(subset=['Waktu'], inplace=True)
data2.dropna(subset=['Waktu'], inplace=True)
data3.dropna(subset=['Waktu'], inplace=True)
data4.dropna(subset=['Date'], inplace=True)

# Set the 'Date' column as the index for all datasets
data1.set_index('Waktu', inplace=True)
data2.set_index('Waktu', inplace=True)
data3.set_index('Waktu', inplace=True)
data4.set_index('Date', inplace=True)

# Select relevant columns from data4 and rename them to match the other datasets
data4 = data4[['PM10', 'SO2', 'CO', 'O3', 'NO2']]
data4['PM2.5'] = np.nan  # Add PM2.5 column with NaN values to match other datasets

# Concatenate the datasets
data = pd.concat([data1, data2, data3, data4])

# Replace commas with periods and '-' with NaN, then convert to float
for col in ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']:
    data[col] = data[col].astype(str).str.replace(',', '.').replace('-', np.nan).astype(float)

# Remove rows with 0 data
data = data[(data[['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']] != 0).all(axis=1)]

# Handle NaN values in the numeric columns only
numeric_columns = ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']
data[numeric_columns] = data[numeric_columns].fillna(method='ffill').fillna(method='bfill')

# Select features and target
features = data[numeric_columns]
target = data[numeric_columns]

# Normalize the features and target
scaler_features = MinMaxScaler()
features_scaled = scaler_features.fit_transform(features)

scaler_target = MinMaxScaler()
target_scaled = scaler_target.fit_transform(target)

# Prepare the data for time series models with sequence length of 5 days
def create_sequences(data, target, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(target[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 5
X, y = create_sequences(features_scaled, target_scaled, seq_length)

# Split the data into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Define a custom attention layer
class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], 1), initializer="normal")
        self.b = self.add_weight(name="att_bias", shape=(1,), initializer="zeros")
        super(Attention, self).build(input_shape)

    def call(self, x):
        e = tf.tanh(tf.linalg.matmul(x, self.W) + self.b)  # Use tf.linalg.matmul
        a = tf.nn.softmax(e, axis=1)  # Use tf.nn.softmax
        output = x * a
        return tf.reduce_sum(output, axis=1)  # Use tf.reduce_sum

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])

# Function to evaluate the model for each pollutant with additional metrics
def evaluate_model_per_pollutant(y_true, y_pred):
    metrics = {}
    for i, pollutant in enumerate(numeric_columns):
        mse = mean_squared_error(y_true[:, i], y_pred[:, i])
        mae = mean_absolute_error(y_true[:, i], y_pred[:, i])
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true[:, i], y_pred[:, i])
        mape = np.mean(np.abs((y_true[:, i] - y_pred[:, i]) / y_true[:, i])) * 100
        mbd = np.mean(y_pred[:, i] - y_true[:, i])
        smape = 100/len(y_true) * np.sum(2 * np.abs(y_pred[:, i] - y_true[:, i]) / (np.abs(y_true[:, i]) + np.abs(y_pred[:, i])))
        metrics[pollutant] = {'MSE': mse, 'MAE': mae, 'RMSE': rmse, 'R-squared': r2, 'MAPE': mape, 'MBD': mbd, 'sMAPE': smape}
    return metrics

# CNN Model
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(seq_length, X.shape[2])))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(y.shape[1]))
cnn_model.compile(optimizer='adam', loss='mean_squared_error')
cnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
cnn_predictions_scaled = cnn_model.predict(X_test)
cnn_predictions = scaler_target.inverse_transform(cnn_predictions_scaled)
cnn_metrics = evaluate_model_per_pollutant(y_test, cnn_predictions_scaled)

# Define CNN model with Attention
def build_cnn_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the CNN model with Attention
cnn_attention_model = build_cnn_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
cnn_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
cnn_attention_predictions_scaled = cnn_attention_model.predict(X_test)
cnn_attention_predictions = scaler_target.inverse_transform(cnn_attention_predictions_scaled)
cnn_attention_metrics = evaluate_model_per_pollutant(y_test, cnn_attention_predictions_scaled)


# RNN Model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
rnn_model.add(SimpleRNN(64))
rnn_model.add(Dense(y.shape[1]))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
rnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
rnn_predictions_scaled = rnn_model.predict(X_test)
rnn_predictions = scaler_target.inverse_transform(rnn_predictions_scaled)
rnn_metrics = evaluate_model_per_pollutant(y_test, rnn_predictions_scaled)
def build_rnn_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = SimpleRNN(256, return_sequences=True)(inputs)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the RNN model with Attention
rnn_attention_model = build_rnn_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
rnn_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
rnn_attention_predictions_scaled = rnn_attention_model.predict(X_test)
rnn_attention_predictions = scaler_target.inverse_transform(rnn_attention_predictions_scaled)
rnn_attention_metrics = evaluate_model_per_pollutant(y_test, rnn_attention_predictions_scaled)

# GRU Model
gru_model = Sequential()
gru_model.add(GRU(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
gru_model.add(GRU(64))
gru_model.add(Dense(y.shape[1]))
gru_model.compile(optimizer='adam', loss='mean_squared_error')
gru_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
gru_predictions_scaled = gru_model.predict(X_test)
gru_predictions = scaler_target.inverse_transform(gru_predictions_scaled)
gru_metrics = evaluate_model_per_pollutant(y_test, gru_predictions_scaled)

# Define GRU model with Attention
def build_gru_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = GRU(256, return_sequences=True)(inputs)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the GRU model with Attention
gru_attention_model = build_gru_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
gru_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
gru_attention_predictions_scaled = gru_attention_model.predict(X_test)
gru_attention_predictions = scaler_target.inverse_transform(gru_attention_predictions_scaled)
gru_attention_metrics = evaluate_model_per_pollutant(y_test, gru_attention_predictions_scaled)

# LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
lstm_model.add(LSTM(64))
lstm_model.add(Dense(y.shape[1]))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
lstm_predictions_scaled = lstm_model.predict(X_test)
lstm_predictions = scaler_target.inverse_transform(lstm_predictions_scaled)
lstm_metrics = evaluate_model_per_pollutant(y_test, lstm_predictions_scaled)
# Define an LSTM model with Attention
def build_lstm_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = LSTM(256, return_sequences=True)(inputs)
    x = Attention()(x)  # Add Attention layer
    x = Dense(output_shape, activation='linear')(x)
    model = Model(inputs=inputs, outputs=x)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
lstm_attention_model = build_lstm_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the LSTM with Attention
lstm_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
lstm_attention_predictions_scaled = lstm_attention_model.predict(X_test)
lstm_attention_predictions = scaler_target.inverse_transform(lstm_attention_predictions_scaled)
lstm_attention_metrics = evaluate_model_per_pollutant(y_test, lstm_attention_predictions_scaled)


# SARIMA Model (only for PM2.5)
sarima_model = SARIMAX(data['PM2.5'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_model.fit(disp=False)
sarima_predictions = sarima_fit.predict(start=len(data) - len(y_test), end=len(data) - 1)
sarima_metrics = {
    'PM2.5': {
        'MSE': mean_squared_error(y_test[:, 0], sarima_predictions),
        'MAE': mean_absolute_error(y_test[:, 0], sarima_predictions),
        'RMSE': np.sqrt(mean_squared_error(y_test[:, 0], sarima_predictions)),
        'R-squared': r2_score(y_test[:, 0], sarima_predictions),
        'MAPE': np.mean(np.abs((y_test[:, 0] - sarima_predictions) / y_test[:, 0])) * 100,
        'MBD': np.mean(sarima_predictions - y_test[:, 0]),
        'sMAPE': 100/len(y_test) * np.sum(2 * np.abs(sarima_predictions - y_test[:, 0]) / (np.abs(y_test[:, 0]) + np.abs(sarima_predictions)))
    }
}

# Define a ResNet block
def resnet_block(input_layer, filters, kernel_size):
    # First convolution
    x = Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(input_layer)
    # Second convolution
    x = Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(x)
    # Add the input (residual connection)
    x = Add()([x, input_layer])
    return x
# Define the ResNet-inspired model
def build_resnet_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    # Add ResNet blocks
    x = resnet_block(x, filters=64, kernel_size=2)
    x = resnet_block(x, filters=64, kernel_size=2)
    # Pooling and flattening
    x = MaxPooling1D(pool_size=2)(x)
    x = Flatten()(x)
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ResNet-inspired model
resnet_model = build_resnet_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ResNet model
resnet_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
resnet_predictions_scaled = resnet_model.predict(X_test)
resnet_predictions = scaler_target.inverse_transform(resnet_predictions_scaled)
resnet_metrics = evaluate_model_per_pollutant(y_test, resnet_predictions_scaled)

# Define the ResNet-inspired model with Attention
def build_resnet_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    # Add ResNet blocks
    x = resnet_block(x, filters=64, kernel_size=2)
    x = resnet_block(x, filters=64, kernel_size=2)
    # Add Attention layer
    x = Attention()(x)
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ResNet-inspired model with Attention
resnet_attention_model = build_resnet_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ResNet model with Attention
resnet_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
resnet_attention_predictions_scaled = resnet_attention_model.predict(X_test)
resnet_attention_predictions = scaler_target.inverse_transform(resnet_attention_predictions_scaled)
resnet_attention_metrics = evaluate_model_per_pollutant(y_test, resnet_attention_predictions_scaled)

# Define a ReXNet block for time series
def rexnet_block(input_layer, filters, expand_ratio):
    # Expand
    x = Conv1D(filters=int(input_layer.shape[-1] * expand_ratio), kernel_size=1, activation=None, padding='same')(input_layer)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    
    # Depthwise convolution
    x = DepthwiseConv1D(kernel_size=3, padding='same', activation=None)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    
    # Project
    x = Conv1D(filters=filters, kernel_size=1, activation=None, padding='same')(x)
    x = BatchNormalization()(x)
    
    # Residual connection
    if int(input_layer.shape[-1]) == filters:
        x = Add()([x, input_layer])
    
    return x
# Define the ReXNet-inspired model
def build_rexnet_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    
    # Initial convolution
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    
    # Add ReXNet blocks
    x = rexnet_block(x, filters=32, expand_ratio=6)
    x = rexnet_block(x, filters=64, expand_ratio=6)
    x = rexnet_block(x, filters=128, expand_ratio=6)
    
    # Global average pooling
    x = GlobalAveragePooling1D()(x)
    
    # Output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ReXNet-inspired model
rexnet_model = build_rexnet_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ReXNet model
rexnet_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
rexnet_predictions_scaled = rexnet_model.predict(X_test)
rexnet_predictions = scaler_target.inverse_transform(rexnet_predictions_scaled)
rexnet_metrics = evaluate_model_per_pollutant(y_test, rexnet_predictions_scaled)

def build_rexnet_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    # Add ReXNet blocks
    x = rexnet_block(x, filters=32, expand_ratio=6)
    x = rexnet_block(x, filters=64, expand_ratio=6)
    x = rexnet_block(x, filters=128, expand_ratio=6) 
    # Add Attention layer
    x = Attention()(x) 
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build the ReXNet-inspired model with Attention
rexnet_attention_model = build_rexnet_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ReXNet model with Attention
rexnet_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
rexnet_attention_predictions_scaled = rexnet_attention_model.predict(X_test)
rexnet_attention_predictions = scaler_target.inverse_transform(rexnet_attention_predictions_scaled)
rexnet_attention_metrics = evaluate_model_per_pollutant(y_test, rexnet_attention_predictions_scaled)


# Print evaluation metrics for each pollutant and model
print("CNN Metrics:")
for pollutant in cnn_metrics:
    print(f"{pollutant} - MSE: {cnn_metrics[pollutant]['MSE']}, MAE: {cnn_metrics[pollutant]['MAE']}, RMSE: {cnn_metrics[pollutant]['RMSE']}, R-squared: {cnn_metrics[pollutant]['R-squared']}, MAPE: {cnn_metrics[pollutant]['MAPE']}, MBD: {cnn_metrics[pollutant]['MBD']}, sMAPE: {cnn_metrics[pollutant]['sMAPE']}")

print("\nCNN with Attention Metrics:")
for pollutant in cnn_attention_metrics:
    print(f"{pollutant} - MSE: {cnn_attention_metrics[pollutant]['MSE']}, MAE: {cnn_attention_metrics[pollutant]['MAE']}, RMSE: {cnn_attention_metrics[pollutant]['RMSE']}, R-squared: {cnn_attention_metrics[pollutant]['R-squared']}, MAPE: {cnn_attention_metrics[pollutant]['MAPE']}, MBD: {cnn_attention_metrics[pollutant]['MBD']}, sMAPE: {cnn_attention_metrics[pollutant]['sMAPE']}")

print("\nRNN Metrics:")
for pollutant in rnn_metrics:
    print(f"{pollutant} - MSE: {rnn_metrics[pollutant]['MSE']}, MAE: {rnn_metrics[pollutant]['MAE']}, RMSE: {rnn_metrics[pollutant]['RMSE']}, R-squared: {rnn_metrics[pollutant]['R-squared']}, MAPE: {rnn_metrics[pollutant]['MAPE']}, MBD: {rnn_metrics[pollutant]['MBD']}, sMAPE: {rnn_metrics[pollutant]['sMAPE']}")

print("\nRNN with Attention Metrics:")
for pollutant in rnn_attention_metrics:
    print(f"{pollutant} - MSE: {rnn_attention_metrics[pollutant]['MSE']}, MAE: {rnn_attention_metrics[pollutant]['MAE']}, RMSE: {rnn_attention_metrics[pollutant]['RMSE']}, R-squared: {rnn_attention_metrics[pollutant]['R-squared']}, MAPE: {rnn_attention_metrics[pollutant]['MAPE']}, MBD: {rnn_attention_metrics[pollutant]['MBD']}, sMAPE: {rnn_attention_metrics[pollutant]['sMAPE']}")

print("\nGRU Metrics:")
for pollutant in gru_metrics:
    print(f"{pollutant} - MSE: {gru_metrics[pollutant]['MSE']}, MAE: {gru_metrics[pollutant]['MAE']}, RMSE: {gru_metrics[pollutant]['RMSE']}, R-squared: {gru_metrics[pollutant]['R-squared']}, MAPE: {gru_metrics[pollutant]['MAPE']}, MBD: {gru_metrics[pollutant]['MBD']}, sMAPE: {gru_metrics[pollutant]['sMAPE']}")

print("\nGRU with Attention Metrics:")
for pollutant in gru_attention_metrics:
    print(f"{pollutant} - MSE: {gru_attention_metrics[pollutant]['MSE']}, MAE: {gru_attention_metrics[pollutant]['MAE']}, RMSE: {gru_attention_metrics[pollutant]['RMSE']}, R-squared: {gru_attention_metrics[pollutant]['R-squared']}, MAPE: {gru_attention_metrics[pollutant]['MAPE']}, MBD: {gru_attention_metrics[pollutant]['MBD']}, sMAPE: {gru_attention_metrics[pollutant]['sMAPE']}")


print("\nLSTM Metrics:")
for pollutant in lstm_metrics:
    print(f"{pollutant} - MSE: {lstm_metrics[pollutant]['MSE']}, MAE: {lstm_metrics[pollutant]['MAE']}, RMSE: {lstm_metrics[pollutant]['RMSE']}, R-squared: {lstm_metrics[pollutant]['R-squared']}, MAPE: {lstm_metrics[pollutant]['MAPE']}, MBD: {lstm_metrics[pollutant]['MBD']}, sMAPE: {lstm_metrics[pollutant]['sMAPE']}")
# Print LSTM with Attention metrics
print("\nLSTM with Attention Metrics:")
for pollutant in lstm_attention_metrics:
    print(f"{pollutant} - MSE: {lstm_attention_metrics[pollutant]['MSE']}, MAE: {lstm_attention_metrics[pollutant]['MAE']}, RMSE: {lstm_attention_metrics[pollutant]['RMSE']}, R-squared: {lstm_attention_metrics[pollutant]['R-squared']}, MAPE: {lstm_attention_metrics[pollutant]['MAPE']}, MBD: {lstm_attention_metrics[pollutant]['MBD']}, sMAPE: {lstm_attention_metrics[pollutant]['sMAPE']}")


print("\nSARIMA Metrics (only for PM2.5):")
for pollutant in sarima_metrics:
    print(f"{pollutant} - MSE: {sarima_metrics[pollutant]['MSE']}, MAE: {sarima_metrics[pollutant]['MAE']}, RMSE: {sarima_metrics[pollutant]['RMSE']}, R-squared: {sarima_metrics[pollutant]['R-squared']}, MAPE: {sarima_metrics[pollutant]['MAPE']}, MBD: {sarima_metrics[pollutant]['MBD']}, sMAPE: {sarima_metrics[pollutant]['sMAPE']}")


# Print ResNet metrics
print("\nResNet Metrics:")
for pollutant in resnet_metrics:
    print(f"{pollutant} - MSE: {resnet_metrics[pollutant]['MSE']}, MAE: {resnet_metrics[pollutant]['MAE']}, RMSE: {resnet_metrics[pollutant]['RMSE']}, R-squared: {resnet_metrics[pollutant]['R-squared']}, MAPE: {resnet_metrics[pollutant]['MAPE']}, MBD: {resnet_metrics[pollutant]['MBD']}, sMAPE: {resnet_metrics[pollutant]['sMAPE']}")

print("\nResNet with Attention Metrics:")
for pollutant in resnet_attention_metrics:
    print(f"{pollutant} - MSE: {resnet_attention_metrics[pollutant]['MSE']}, MAE: {resnet_attention_metrics[pollutant]['MAE']}, RMSE: {resnet_attention_metrics[pollutant]['RMSE']}, R-squared: {resnet_attention_metrics[pollutant]['R-squared']}, MAPE: {resnet_attention_metrics[pollutant]['MAPE']}, MBD: {resnet_attention_metrics[pollutant]['MBD']}, sMAPE: {resnet_attention_metrics[pollutant]['sMAPE']}")


# Print ReXNet metrics
print("\nReXNet Metrics:")
for pollutant in rexnet_metrics:
    print(f"{pollutant} - MSE: {rexnet_metrics[pollutant]['MSE']}, MAE: {rexnet_metrics[pollutant]['MAE']}, RMSE: {rexnet_metrics[pollutant]['RMSE']}, R-squared: {rexnet_metrics[pollutant]['R-squared']}, MAPE: {rexnet_metrics[pollutant]['MAPE']}, MBD: {rexnet_metrics[pollutant]['MBD']}, sMAPE: {rexnet_metrics[pollutant]['sMAPE']}")

print("\nReXNet with Attention Metrics:")
for pollutant in rexnet_attention_metrics:
    print(f"{pollutant} - MSE: {rexnet_attention_metrics[pollutant]['MSE']}, MAE: {rexnet_attention_metrics[pollutant]['MAE']}, RMSE: {rexnet_attention_metrics[pollutant]['RMSE']}, R-squared: {rexnet_attention_metrics[pollutant]['R-squared']}, MAPE: {rexnet_attention_metrics[pollutant]['MAPE']}, MBD: {rexnet_attention_metrics[pollutant]['MBD']}, sMAPE: {rexnet_attention_metrics[pollutant]['sMAPE']}")


# Plot each pollutant over time and compare actual vs predicted values for test set
pollutants = ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']
for pollutant in pollutants:
    plt.figure(figsize=(12, 6))
    plt.plot(data.index[-len(X_test):], scaler_target.inverse_transform(y_test)[:, pollutants.index(pollutant)], label=f'Actual {pollutant}')
    plt.plot(data.index[-len(X_test):], cnn_predictions[:, pollutants.index(pollutant)], label=f'CNN Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], cnn_attention_predictions[:, pollutants.index(pollutant)], label=f'CNN + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rnn_predictions[:, pollutants.index(pollutant)], label=f'RNN Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rnn_attention_predictions[:, pollutants.index(pollutant)], label=f'RNN + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], gru_predictions[:, pollutants.index(pollutant)], label=f'GRU Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], gru_attention_predictions[:, pollutants.index(pollutant)], label=f'GRU + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], lstm_predictions[:, pollutants.index(pollutant)], label=f'LSTM Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], lstm_attention_predictions[:, pollutants.index(pollutant)], label=f'LSTM + Attention Predicted {pollutant}')
    if pollutant == 'PM2.5':
        plt.plot(data.index[-len(X_test):], sarima_predictions, label=f'SARIMA Predicted {pollutant}')
    #add resnet,rexnet
    plt.plot(data.index[-len(X_test):], resnet_predictions[:, pollutants.index(pollutant)], label=f'ResNet Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], resnet_attention_predictions[:, pollutants.index(pollutant)], label=f'ResNet + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rexnet_predictions[:, pollutants.index(pollutant)], label=f'ReXNet Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rexnet_attention_predictions[:, pollutants.index(pollutant)], label=f'ReXNet + Attention Predicted {pollutant}')
    
    plt.title(f'{pollutant} over Time (Actual vs Predicted)')
    plt.xlabel('Date')
    plt.ylabel(pollutant)
    plt.legend()
    plt.show()

# Ensemble
# Step 1: Collect all predictions into a dictionary
all_predictions = {
    "CNN": cnn_predictions,
    "CNN + Attention": cnn_attention_predictions,
    "RNN": rnn_predictions,
    "RNN + Attention": rnn_attention_predictions,
    "GRU": gru_predictions,
    "GRU + Attention": gru_attention_predictions,
    "LSTM": lstm_predictions,
    "LSTM + Attention": lstm_attention_predictions,
    "ResNet": resnet_predictions,
    "ResNet + Attention": resnet_attention_predictions,
    "ReXNet": rexnet_predictions,
    "ReXNet + Attention": rexnet_attention_predictions,
}

# Add SARIMA predictions (only for PM2.5)
all_predictions["SARIMA"] = np.zeros_like(cnn_predictions)
all_predictions["SARIMA"][:, 0] = sarima_predictions  # Only for PM2.5

# Convert predictions into a NumPy array for easier processing
predictions_array = np.array(list(all_predictions.values()))  # Shape: (num_models, num_samples, num_features)

# Step 2: Define the objective function to minimize (e.g., weighted MSE)
def weighted_mse(weights):
    # Compute the weighted ensemble predictions
    ensemble_predictions = np.tensordot(weights, predictions_array, axes=(0, 0))  # Weighted sum along model axis
    # Compute the MSE between the ensemble predictions and true values
    return mean_squared_error(y_test, ensemble_predictions)

# Step 3: Initial weights (equal weights for all models)
num_models = predictions_array.shape[0]
initial_weights = np.ones(num_models) / num_models

# Step 4: Define constraints (weights must sum to 1) and bounds (weights between 0 and 1)
constraints = {"type": "eq", "fun": lambda w: np.sum(w) - 1}  # Weights sum to 1
bounds = [(0, 1)] * num_models  # Weights between 0 and 1

# Step 5: Optimize the weights using scipy
result = minimize(weighted_mse, initial_weights, bounds=bounds, constraints=constraints)

# Step 6: Extract optimized weights
optimized_weights = result.x
print(f"Optimized Weights: {optimized_weights}")

# Step 7: Compute final ensemble predictions using optimized weights
optimized_ensemble_predictions = np.tensordot(optimized_weights, predictions_array, axes=(0, 0))

# Step 8: Evaluate the optimized ensemble
optimized_ensemble_metrics = evaluate_model_per_pollutant(y_test, optimized_ensemble_predictions)
print("\nOptimized Weighted Ensemble Metrics:")
for pollutant in optimized_ensemble_metrics:
    print(f"{pollutant} - MSE: {optimized_ensemble_metrics[pollutant]['MSE']}, "
          f"MAE: {optimized_ensemble_metrics[pollutant]['MAE']}, RMSE: {optimized_ensemble_metrics[pollutant]['RMSE']}, "
          f"R-squared: {optimized_ensemble_metrics[pollutant]['R-squared']}")

# Plot the optimized ensemble predictions
for pollutant in pollutants:
    plt.figure(figsize=(12, 6))
    plt.plot(data.index[-len(X_test):], scaler_target.inverse_transform(y_test)[:, pollutants.index(pollutant)], label=f'Actual {pollutant}')
    plt.plot(data.index[-len(X_test):], optimized_ensemble_predictions[:, pollutants.index(pollutant)], label=f'Optimized Weighted Ensemble {pollutant}')
    plt.title(f'{pollutant} - Actual vs Optimized Weighted Ensemble')
    plt.xlabel('Date')
    plt.ylabel(pollutant)
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from keras.models import Sequential, Model
from keras.layers import Input, Add, Conv1D, MaxPooling1D, Flatten, LSTM, GRU, Dense, SimpleRNN,ReLU,DepthwiseConv1D,GlobalAveragePooling1D,BatchNormalization,Layer
from statsmodels.tsa.statespace.sarimax import SARIMAX
import keras.backend as K
import tensorflow as tf
from scipy.optimize import minimize

# Load the dataset
data = pd.read_csv('2020-2022_tangsel_cleaned.csv', delimiter=';')

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y', errors='coerce')

# Drop rows with invalid dates
data.dropna(subset=['Date'], inplace=True)

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# Replace commas with periods and '-' with NaN, then convert to float
for col in ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']:
    data[col] = data[col].astype(str).str.replace(',', '.').replace('-', np.nan).astype(float)

# Remove rows with 0 data
data = data[(data[['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']] != 0).all(axis=1)]

# Handle NaN values in the numeric columns only
numeric_columns = ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']
data[numeric_columns] = data[numeric_columns].fillna(method='ffill').fillna(method='bfill')

# Select features and target
features = data[numeric_columns]
target = data[numeric_columns]

# Normalize the features and target
scaler_features = MinMaxScaler()
features_scaled = scaler_features.fit_transform(features)

scaler_target = MinMaxScaler()
target_scaled = scaler_target.fit_transform(target)

# Prepare the data for time series models with sequence length of 5 days
def create_sequences(data, target, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(target[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 5
X, y = create_sequences(features_scaled, target_scaled, seq_length)

# Split the data into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Define a custom attention layer
class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], 1), initializer="normal")
        self.b = self.add_weight(name="att_bias", shape=(1,), initializer="zeros")
        super(Attention, self).build(input_shape)

    def call(self, x):
        e = tf.tanh(tf.linalg.matmul(x, self.W) + self.b)  # Use tf.linalg.matmul
        a = tf.nn.softmax(e, axis=1)  # Use tf.nn.softmax
        output = x * a
        return tf.reduce_sum(output, axis=1)  # Use tf.reduce_sum

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])

# Function to evaluate the model for each pollutant with additional metrics
def evaluate_model_per_pollutant(y_true, y_pred):
    metrics = {}
    for i, pollutant in enumerate(numeric_columns):
        mse = mean_squared_error(y_true[:, i], y_pred[:, i])
        mae = mean_absolute_error(y_true[:, i], y_pred[:, i])
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true[:, i], y_pred[:, i])
        mape = np.mean(np.abs((y_true[:, i] - y_pred[:, i]) / y_true[:, i])) * 100
        mbd = np.mean(y_pred[:, i] - y_true[:, i])
        smape = 100/len(y_true) * np.sum(2 * np.abs(y_pred[:, i] - y_true[:, i]) / (np.abs(y_true[:, i]) + np.abs(y_pred[:, i])))
        metrics[pollutant] = {'MSE': mse, 'MAE': mae, 'RMSE': rmse, 'R-squared': r2, 'MAPE': mape, 'MBD': mbd, 'sMAPE': smape}
    return metrics

# CNN Model
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(seq_length, X.shape[2])))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(y.shape[1]))
cnn_model.compile(optimizer='adam', loss='mean_squared_error')
cnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
cnn_predictions_scaled = cnn_model.predict(X_test)
cnn_predictions = scaler_target.inverse_transform(cnn_predictions_scaled)
cnn_metrics = evaluate_model_per_pollutant(y_test, cnn_predictions_scaled)

# Define CNN model with Attention
def build_cnn_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the CNN model with Attention
cnn_attention_model = build_cnn_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
cnn_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
cnn_attention_predictions_scaled = cnn_attention_model.predict(X_test)
cnn_attention_predictions = scaler_target.inverse_transform(cnn_attention_predictions_scaled)
cnn_attention_metrics = evaluate_model_per_pollutant(y_test, cnn_attention_predictions_scaled)


# RNN Model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
rnn_model.add(SimpleRNN(64))
rnn_model.add(Dense(y.shape[1]))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
rnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
rnn_predictions_scaled = rnn_model.predict(X_test)
rnn_predictions = scaler_target.inverse_transform(rnn_predictions_scaled)
rnn_metrics = evaluate_model_per_pollutant(y_test, rnn_predictions_scaled)
def build_rnn_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = SimpleRNN(256, return_sequences=True)(inputs)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the RNN model with Attention
rnn_attention_model = build_rnn_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
rnn_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
rnn_attention_predictions_scaled = rnn_attention_model.predict(X_test)
rnn_attention_predictions = scaler_target.inverse_transform(rnn_attention_predictions_scaled)
rnn_attention_metrics = evaluate_model_per_pollutant(y_test, rnn_attention_predictions_scaled)

# GRU Model
gru_model = Sequential()
gru_model.add(GRU(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
gru_model.add(GRU(64))
gru_model.add(Dense(y.shape[1]))
gru_model.compile(optimizer='adam', loss='mean_squared_error')
gru_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
gru_predictions_scaled = gru_model.predict(X_test)
gru_predictions = scaler_target.inverse_transform(gru_predictions_scaled)
gru_metrics = evaluate_model_per_pollutant(y_test, gru_predictions_scaled)

# Define GRU model with Attention
def build_gru_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = GRU(256, return_sequences=True)(inputs)
    
    # Add Attention layer
    x = Attention()(x)
    
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build and train the GRU model with Attention
gru_attention_model = build_gru_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
gru_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
gru_attention_predictions_scaled = gru_attention_model.predict(X_test)
gru_attention_predictions = scaler_target.inverse_transform(gru_attention_predictions_scaled)
gru_attention_metrics = evaluate_model_per_pollutant(y_test, gru_attention_predictions_scaled)

# LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(256, return_sequences=True, input_shape=(seq_length, X.shape[2])))
lstm_model.add(LSTM(64))
lstm_model.add(Dense(y.shape[1]))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
lstm_predictions_scaled = lstm_model.predict(X_test)
lstm_predictions = scaler_target.inverse_transform(lstm_predictions_scaled)
lstm_metrics = evaluate_model_per_pollutant(y_test, lstm_predictions_scaled)
# Define an LSTM model with Attention
def build_lstm_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    x = LSTM(256, return_sequences=True)(inputs)
    x = Attention()(x)  # Add Attention layer
    x = Dense(output_shape, activation='linear')(x)
    model = Model(inputs=inputs, outputs=x)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
lstm_attention_model = build_lstm_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the LSTM with Attention
lstm_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
lstm_attention_predictions_scaled = lstm_attention_model.predict(X_test)
lstm_attention_predictions = scaler_target.inverse_transform(lstm_attention_predictions_scaled)
lstm_attention_metrics = evaluate_model_per_pollutant(y_test, lstm_attention_predictions_scaled)


# SARIMA Model (only for PM2.5)
sarima_model = SARIMAX(data['PM2.5'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_model.fit(disp=False)
sarima_predictions = sarima_fit.predict(start=len(data) - len(y_test), end=len(data) - 1)
sarima_metrics = {
    'PM2.5': {
        'MSE': mean_squared_error(y_test[:, 0], sarima_predictions),
        'MAE': mean_absolute_error(y_test[:, 0], sarima_predictions),
        'RMSE': np.sqrt(mean_squared_error(y_test[:, 0], sarima_predictions)),
        'R-squared': r2_score(y_test[:, 0], sarima_predictions),
        'MAPE': np.mean(np.abs((y_test[:, 0] - sarima_predictions) / y_test[:, 0])) * 100,
        'MBD': np.mean(sarima_predictions - y_test[:, 0]),
        'sMAPE': 100/len(y_test) * np.sum(2 * np.abs(sarima_predictions - y_test[:, 0]) / (np.abs(y_test[:, 0]) + np.abs(sarima_predictions)))
    }
}

# Define a ResNet block
def resnet_block(input_layer, filters, kernel_size):
    # First convolution
    x = Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(input_layer)
    # Second convolution
    x = Conv1D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(x)
    # Add the input (residual connection)
    x = Add()([x, input_layer])
    return x
# Define the ResNet-inspired model
def build_resnet_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    # Add ResNet blocks
    x = resnet_block(x, filters=64, kernel_size=2)
    x = resnet_block(x, filters=64, kernel_size=2)
    # Pooling and flattening
    x = MaxPooling1D(pool_size=2)(x)
    x = Flatten()(x)
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ResNet-inspired model
resnet_model = build_resnet_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ResNet model
resnet_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
resnet_predictions_scaled = resnet_model.predict(X_test)
resnet_predictions = scaler_target.inverse_transform(resnet_predictions_scaled)
resnet_metrics = evaluate_model_per_pollutant(y_test, resnet_predictions_scaled)

# Define the ResNet-inspired model with Attention
def build_resnet_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=64, kernel_size=2, activation='relu', padding='same')(inputs)
    # Add ResNet blocks
    x = resnet_block(x, filters=64, kernel_size=2)
    x = resnet_block(x, filters=64, kernel_size=2)
    # Add Attention layer
    x = Attention()(x)
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ResNet-inspired model with Attention
resnet_attention_model = build_resnet_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ResNet model with Attention
resnet_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
resnet_attention_predictions_scaled = resnet_attention_model.predict(X_test)
resnet_attention_predictions = scaler_target.inverse_transform(resnet_attention_predictions_scaled)
resnet_attention_metrics = evaluate_model_per_pollutant(y_test, resnet_attention_predictions_scaled)

# Define a ReXNet block for time series
def rexnet_block(input_layer, filters, expand_ratio):
    # Expand
    x = Conv1D(filters=int(input_layer.shape[-1] * expand_ratio), kernel_size=1, activation=None, padding='same')(input_layer)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    
    # Depthwise convolution
    x = DepthwiseConv1D(kernel_size=3, padding='same', activation=None)(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    
    # Project
    x = Conv1D(filters=filters, kernel_size=1, activation=None, padding='same')(x)
    x = BatchNormalization()(x)
    
    # Residual connection
    if int(input_layer.shape[-1]) == filters:
        x = Add()([x, input_layer])
    
    return x
# Define the ReXNet-inspired model
def build_rexnet_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    
    # Initial convolution
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    
    # Add ReXNet blocks
    x = rexnet_block(x, filters=32, expand_ratio=6)
    x = rexnet_block(x, filters=64, expand_ratio=6)
    x = rexnet_block(x, filters=128, expand_ratio=6)
    
    # Global average pooling
    x = GlobalAveragePooling1D()(x)
    
    # Output layer
    outputs = Dense(output_shape, activation='linear')(x)
    
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
# Build the ReXNet-inspired model
rexnet_model = build_rexnet_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ReXNet model
rexnet_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
# Predict and evaluate
rexnet_predictions_scaled = rexnet_model.predict(X_test)
rexnet_predictions = scaler_target.inverse_transform(rexnet_predictions_scaled)
rexnet_metrics = evaluate_model_per_pollutant(y_test, rexnet_predictions_scaled)

def build_rexnet_attention_model(input_shape, output_shape):
    inputs = Input(shape=input_shape)
    # Initial convolution
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    # Add ReXNet blocks
    x = rexnet_block(x, filters=32, expand_ratio=6)
    x = rexnet_block(x, filters=64, expand_ratio=6)
    x = rexnet_block(x, filters=128, expand_ratio=6) 
    # Add Attention layer
    x = Attention()(x) 
    # Dense output layer
    outputs = Dense(output_shape, activation='linear')(x)
    # Create the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Build the ReXNet-inspired model with Attention
rexnet_attention_model = build_rexnet_attention_model(input_shape=(seq_length, X.shape[2]), output_shape=y.shape[1])
# Train the ReXNet model with Attention
rexnet_attention_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
rexnet_attention_predictions_scaled = rexnet_attention_model.predict(X_test)
rexnet_attention_predictions = scaler_target.inverse_transform(rexnet_attention_predictions_scaled)
rexnet_attention_metrics = evaluate_model_per_pollutant(y_test, rexnet_attention_predictions_scaled)


# Print evaluation metrics for each pollutant and model
print("CNN Metrics:")
for pollutant in cnn_metrics:
    print(f"{pollutant} - MSE: {cnn_metrics[pollutant]['MSE']}, MAE: {cnn_metrics[pollutant]['MAE']}, RMSE: {cnn_metrics[pollutant]['RMSE']}, R-squared: {cnn_metrics[pollutant]['R-squared']}, MAPE: {cnn_metrics[pollutant]['MAPE']}, MBD: {cnn_metrics[pollutant]['MBD']}, sMAPE: {cnn_metrics[pollutant]['sMAPE']}")

print("\nCNN with Attention Metrics:")
for pollutant in cnn_attention_metrics:
    print(f"{pollutant} - MSE: {cnn_attention_metrics[pollutant]['MSE']}, MAE: {cnn_attention_metrics[pollutant]['MAE']}, RMSE: {cnn_attention_metrics[pollutant]['RMSE']}, R-squared: {cnn_attention_metrics[pollutant]['R-squared']}, MAPE: {cnn_attention_metrics[pollutant]['MAPE']}, MBD: {cnn_attention_metrics[pollutant]['MBD']}, sMAPE: {cnn_attention_metrics[pollutant]['sMAPE']}")

print("\nRNN Metrics:")
for pollutant in rnn_metrics:
    print(f"{pollutant} - MSE: {rnn_metrics[pollutant]['MSE']}, MAE: {rnn_metrics[pollutant]['MAE']}, RMSE: {rnn_metrics[pollutant]['RMSE']}, R-squared: {rnn_metrics[pollutant]['R-squared']}, MAPE: {rnn_metrics[pollutant]['MAPE']}, MBD: {rnn_metrics[pollutant]['MBD']}, sMAPE: {rnn_metrics[pollutant]['sMAPE']}")

print("\nRNN with Attention Metrics:")
for pollutant in rnn_attention_metrics:
    print(f"{pollutant} - MSE: {rnn_attention_metrics[pollutant]['MSE']}, MAE: {rnn_attention_metrics[pollutant]['MAE']}, RMSE: {rnn_attention_metrics[pollutant]['RMSE']}, R-squared: {rnn_attention_metrics[pollutant]['R-squared']}, MAPE: {rnn_attention_metrics[pollutant]['MAPE']}, MBD: {rnn_attention_metrics[pollutant]['MBD']}, sMAPE: {rnn_attention_metrics[pollutant]['sMAPE']}")

print("\nGRU Metrics:")
for pollutant in gru_metrics:
    print(f"{pollutant} - MSE: {gru_metrics[pollutant]['MSE']}, MAE: {gru_metrics[pollutant]['MAE']}, RMSE: {gru_metrics[pollutant]['RMSE']}, R-squared: {gru_metrics[pollutant]['R-squared']}, MAPE: {gru_metrics[pollutant]['MAPE']}, MBD: {gru_metrics[pollutant]['MBD']}, sMAPE: {gru_metrics[pollutant]['sMAPE']}")

print("\nGRU with Attention Metrics:")
for pollutant in gru_attention_metrics:
    print(f"{pollutant} - MSE: {gru_attention_metrics[pollutant]['MSE']}, MAE: {gru_attention_metrics[pollutant]['MAE']}, RMSE: {gru_attention_metrics[pollutant]['RMSE']}, R-squared: {gru_attention_metrics[pollutant]['R-squared']}, MAPE: {gru_attention_metrics[pollutant]['MAPE']}, MBD: {gru_attention_metrics[pollutant]['MBD']}, sMAPE: {gru_attention_metrics[pollutant]['sMAPE']}")


print("\nLSTM Metrics:")
for pollutant in lstm_metrics:
    print(f"{pollutant} - MSE: {lstm_metrics[pollutant]['MSE']}, MAE: {lstm_metrics[pollutant]['MAE']}, RMSE: {lstm_metrics[pollutant]['RMSE']}, R-squared: {lstm_metrics[pollutant]['R-squared']}, MAPE: {lstm_metrics[pollutant]['MAPE']}, MBD: {lstm_metrics[pollutant]['MBD']}, sMAPE: {lstm_metrics[pollutant]['sMAPE']}")
# Print LSTM with Attention metrics
print("\nLSTM with Attention Metrics:")
for pollutant in lstm_attention_metrics:
    print(f"{pollutant} - MSE: {lstm_attention_metrics[pollutant]['MSE']}, MAE: {lstm_attention_metrics[pollutant]['MAE']}, RMSE: {lstm_attention_metrics[pollutant]['RMSE']}, R-squared: {lstm_attention_metrics[pollutant]['R-squared']}, MAPE: {lstm_attention_metrics[pollutant]['MAPE']}, MBD: {lstm_attention_metrics[pollutant]['MBD']}, sMAPE: {lstm_attention_metrics[pollutant]['sMAPE']}")


print("\nSARIMA Metrics (only for PM2.5):")
for pollutant in sarima_metrics:
    print(f"{pollutant} - MSE: {sarima_metrics[pollutant]['MSE']}, MAE: {sarima_metrics[pollutant]['MAE']}, RMSE: {sarima_metrics[pollutant]['RMSE']}, R-squared: {sarima_metrics[pollutant]['R-squared']}, MAPE: {sarima_metrics[pollutant]['MAPE']}, MBD: {sarima_metrics[pollutant]['MBD']}, sMAPE: {sarima_metrics[pollutant]['sMAPE']}")


# Print ResNet metrics
print("\nResNet Metrics:")
for pollutant in resnet_metrics:
    print(f"{pollutant} - MSE: {resnet_metrics[pollutant]['MSE']}, MAE: {resnet_metrics[pollutant]['MAE']}, RMSE: {resnet_metrics[pollutant]['RMSE']}, R-squared: {resnet_metrics[pollutant]['R-squared']}, MAPE: {resnet_metrics[pollutant]['MAPE']}, MBD: {resnet_metrics[pollutant]['MBD']}, sMAPE: {resnet_metrics[pollutant]['sMAPE']}")

print("\nResNet with Attention Metrics:")
for pollutant in resnet_attention_metrics:
    print(f"{pollutant} - MSE: {resnet_attention_metrics[pollutant]['MSE']}, MAE: {resnet_attention_metrics[pollutant]['MAE']}, RMSE: {resnet_attention_metrics[pollutant]['RMSE']}, R-squared: {resnet_attention_metrics[pollutant]['R-squared']}, MAPE: {resnet_attention_metrics[pollutant]['MAPE']}, MBD: {resnet_attention_metrics[pollutant]['MBD']}, sMAPE: {resnet_attention_metrics[pollutant]['sMAPE']}")


# Print ReXNet metrics
print("\nReXNet Metrics:")
for pollutant in rexnet_metrics:
    print(f"{pollutant} - MSE: {rexnet_metrics[pollutant]['MSE']}, MAE: {rexnet_metrics[pollutant]['MAE']}, RMSE: {rexnet_metrics[pollutant]['RMSE']}, R-squared: {rexnet_metrics[pollutant]['R-squared']}, MAPE: {rexnet_metrics[pollutant]['MAPE']}, MBD: {rexnet_metrics[pollutant]['MBD']}, sMAPE: {rexnet_metrics[pollutant]['sMAPE']}")

print("\nReXNet with Attention Metrics:")
for pollutant in rexnet_attention_metrics:
    print(f"{pollutant} - MSE: {rexnet_attention_metrics[pollutant]['MSE']}, MAE: {rexnet_attention_metrics[pollutant]['MAE']}, RMSE: {rexnet_attention_metrics[pollutant]['RMSE']}, R-squared: {rexnet_attention_metrics[pollutant]['R-squared']}, MAPE: {rexnet_attention_metrics[pollutant]['MAPE']}, MBD: {rexnet_attention_metrics[pollutant]['MBD']}, sMAPE: {rexnet_attention_metrics[pollutant]['sMAPE']}")


# Plot each pollutant over time and compare actual vs predicted values for test set
pollutants = ['PM2.5', 'PM10', 'SO2', 'CO', 'O3', 'NO2']
for pollutant in pollutants:
    plt.figure(figsize=(12, 6))
    plt.plot(data.index[-len(X_test):], scaler_target.inverse_transform(y_test)[:, pollutants.index(pollutant)], label=f'Actual {pollutant}')
    plt.plot(data.index[-len(X_test):], cnn_predictions[:, pollutants.index(pollutant)], label=f'CNN Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], cnn_attention_predictions[:, pollutants.index(pollutant)], label=f'CNN + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rnn_predictions[:, pollutants.index(pollutant)], label=f'RNN Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rnn_attention_predictions[:, pollutants.index(pollutant)], label=f'RNN + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], gru_predictions[:, pollutants.index(pollutant)], label=f'GRU Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], gru_attention_predictions[:, pollutants.index(pollutant)], label=f'GRU + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], lstm_predictions[:, pollutants.index(pollutant)], label=f'LSTM Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], lstm_attention_predictions[:, pollutants.index(pollutant)], label=f'LSTM + Attention Predicted {pollutant}')
    if pollutant == 'PM2.5':
        plt.plot(data.index[-len(X_test):], sarima_predictions, label=f'SARIMA Predicted {pollutant}')
    #add resnet,rexnet
    plt.plot(data.index[-len(X_test):], resnet_predictions[:, pollutants.index(pollutant)], label=f'ResNet Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], resnet_attention_predictions[:, pollutants.index(pollutant)], label=f'ResNet + Attention Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rexnet_predictions[:, pollutants.index(pollutant)], label=f'ReXNet Predicted {pollutant}')
    plt.plot(data.index[-len(X_test):], rexnet_attention_predictions[:, pollutants.index(pollutant)], label=f'ReXNet + Attention Predicted {pollutant}')
    
    plt.title(f'{pollutant} over Time (Actual vs Predicted)')
    plt.xlabel('Date')
    plt.ylabel(pollutant)
    plt.legend()
    plt.show()

# Ensemble
# Step 1: Collect all predictions into a dictionary
all_predictions = {
    "CNN": cnn_predictions,
    "CNN + Attention": cnn_attention_predictions,
    "RNN": rnn_predictions,
    "RNN + Attention": rnn_attention_predictions,
    "GRU": gru_predictions,
    "GRU + Attention": gru_attention_predictions,
    "LSTM": lstm_predictions,
    "LSTM + Attention": lstm_attention_predictions,
    "ResNet": resnet_predictions,
    "ResNet + Attention": resnet_attention_predictions,
    "ReXNet": rexnet_predictions,
    "ReXNet + Attention": rexnet_attention_predictions,
}

# Add SARIMA predictions (only for PM2.5)
all_predictions["SARIMA"] = np.zeros_like(cnn_predictions)
all_predictions["SARIMA"][:, 0] = sarima_predictions  # Only for PM2.5

# Convert predictions into a NumPy array for easier processing
predictions_array = np.array(list(all_predictions.values()))  # Shape: (num_models, num_samples, num_features)

# Step 2: Define the objective function to minimize (e.g., weighted MSE)
def weighted_mse(weights):
    # Compute the weighted ensemble predictions
    ensemble_predictions = np.tensordot(weights, predictions_array, axes=(0, 0))  # Weighted sum along model axis
    # Compute the MSE between the ensemble predictions and true values
    return mean_squared_error(y_test, ensemble_predictions)

# Step 3: Initial weights (equal weights for all models)
num_models = predictions_array.shape[0]
initial_weights = np.ones(num_models) / num_models

# Step 4: Define constraints (weights must sum to 1) and bounds (weights between 0 and 1)
constraints = {"type": "eq", "fun": lambda w: np.sum(w) - 1}  # Weights sum to 1
bounds = [(0, 1)] * num_models  # Weights between 0 and 1

# Step 5: Optimize the weights using scipy
result = minimize(weighted_mse, initial_weights, bounds=bounds, constraints=constraints)

# Step 6: Extract optimized weights
optimized_weights = result.x
print(f"Optimized Weights: {optimized_weights}")

# Step 7: Compute final ensemble predictions using optimized weights
optimized_ensemble_predictions = np.tensordot(optimized_weights, predictions_array, axes=(0, 0))

# Step 8: Evaluate the optimized ensemble
optimized_ensemble_metrics = evaluate_model_per_pollutant(y_test, optimized_ensemble_predictions)
print("\nOptimized Weighted Ensemble Metrics:")
for pollutant in optimized_ensemble_metrics:
    print(f"{pollutant} - MSE: {optimized_ensemble_metrics[pollutant]['MSE']}, "
          f"MAE: {optimized_ensemble_metrics[pollutant]['MAE']}, RMSE: {optimized_ensemble_metrics[pollutant]['RMSE']}, "
          f"R-squared: {optimized_ensemble_metrics[pollutant]['R-squared']}")

# Plot the optimized ensemble predictions
for pollutant in pollutants:
    plt.figure(figsize=(12, 6))
    plt.plot(data.index[-len(X_test):], scaler_target.inverse_transform(y_test)[:, pollutants.index(pollutant)], label=f'Actual {pollutant}')
    plt.plot(data.index[-len(X_test):], optimized_ensemble_predictions[:, pollutants.index(pollutant)], label=f'Optimized Weighted Ensemble {pollutant}')
    plt.title(f'{pollutant} - Actual vs Optimized Weighted Ensemble')
    plt.xlabel('Date')
    plt.ylabel(pollutant)
    plt.legend()
    plt.show()
```
